```python
from astropy.io import fits
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
import matplotlib
params={
    'text.usetex':True,
    'font.family':'serif',
    'xtick.minor.visible':True,
    'ytick.minor.visible':True,
    'xtick.top':True,
    'ytick.left':True,
    'ytick.right':True,
    'xtick.direction':'out',
    'ytick.direction':'out',
    'xtick.minor.size':2.5,
    'xtick.major.size':5,
    'ytick.minor.size':2.5,
    'ytick.major.size':5,
    'axes.axisbelow':True
}
matplotlib.rcParams.update(params)
```

Open the clean lightcurve files

```python
lc1 = fits.open('AS1A06_002T02_9000003206_21631cztM0_level2_quad_clean_Q0.lc')
lc2 = fits.open('AS1A06_002T02_9000003206_21631cztM0_level2_quad_clean_Q1.lc')
lc3 = fits.open('AS1A06_002T02_9000003206_21631cztM0_level2_quad_clean_Q2.lc')
lc4 = fits.open('AS1A06_002T02_9000003206_21631cztM0_level2_quad_clean_Q3.lc')
```

Extract the data from clean and unclean files

```python
lc1_data = lc1[1].data
lc2_data = lc2[1].data
lc3_data = lc3[1].data
lc4_data = lc4[1].data
```

```python
clean_lc = [lc1_data, lc2_data, lc3_data, lc4_data]
```

Plot the count vs time for all four quadrants

```python
for i in range (0,4):
    plt.figure(dpi=250, figsize=(15,5))
    plt.suptitle('Q'+str(i)+ ' lightcurve')
    plt.xlabel('Time')
    plt.ylabel('Counts/s')
    plt.plot(clean_lc[i].field('TIME'), clean_lc[i].field('RATE'))
    
    plt.savefig('Q'+str(i)+' lightcurve.png')
```

```python
evt_file_1 = fits.open('AS1A06_002T02_9000003206_21631cztM0_level2_quad_clean_Q0.pha')
evt_file_2 = fits.open('AS1A06_002T02_9000003206_21631cztM0_level2_quad_clean_Q1.pha')
evt_file_3 = fits.open('AS1A06_002T02_9000003206_21631cztM0_level2_quad_clean_Q2.pha')
evt_file_4 = fits.open('AS1A06_002T02_9000003206_21631cztM0_level2_quad_clean_Q3.pha')
```

```python
specdata1 = evt_file_1[1].data
specdata2 = evt_file_2[1].data
specdata3 = evt_file_3[1].data
specdata4 = evt_file_4[1].data
```

```python
plt.figure(dpi=500, figsize=(15,6))
plt.title("Spectral Plot")
plt.plot(specdata1['CHANNEL'], specdata1['COUNTS'])
plt.plot(specdata2['CHANNEL'], specdata2['COUNTS'])
plt.plot(specdata3['CHANNEL'], specdata3['COUNTS'])
plt.plot(specdata4['CHANNEL'], specdata4['COUNTS'])
plt.xlabel('Energy (keV)')
plt.ylabel('Counts')
plt.legend(['Quadrant 0','Quadrant 1','Quadrant 2', 'Quadrant 3'])
plt.show()
```
